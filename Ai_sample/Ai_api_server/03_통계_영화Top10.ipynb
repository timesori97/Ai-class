```python
## top10을 선정할 때 vote_count와 vote_average 컬럼으로 선정한다.
## kaggle을 참조하여 공식을 이용한다.
# vote_average의 평균
C = top10_data['vote_average'].mean(axis = 0)

# vote_count의 10% 지점의 값
m = top10_data['vote_count'].quantile(0.9)

print(C, m) 

# 상위 10%만 참여한 별점만 인정한다.
a = top10_data['vote_count'] >= m
a # 1803개의 row

tt = top10_data.loc[a]
tt # 481개의 row가 됨
```

```python
##위에서 만든 평균,10프로시점의 인원수를 가지고 최종 score를 선정하는 공식을 정의합니다
def score_rating(x,m=m,c=C): #x는 각 로우라고 생각하세요 
    v = x['vote_count'] #x의 로우가 통째로 들어오면 그 중에 vote_count값을 v에 저장하시오
    R = x['vote_average']
    return (v/(v+m)*R)+(m/(m+v)*C)
    
```

```python
temp_score = tt.apply(score_rating, axis=1) #apply는 top10_data 로우별로 각각 적용시켜라 이때 axis=1 지정하여 행을 
                                                    #중심으로 각각 적용시켜라
temp_score          ## 총 481개인데, 인덱스는 순차적으로 증가하는 게 아닙니다. 위에서 10프로 선정한 결과의 인덱스입니다          
```

```python
tt['score']=temp_score
tt
```

```python
## tt 데이터프레임에서 score를 기준으로 내림차순 정렬하면 가장 높은 점수부터 정렬된다
tt = tt.sort_values('score',ascending=False)
tt
```
